# Week 2 - Client Requests Statistical Analysis

This notebook solves all 10 client requests from `client_requests.pdf` using `datasets/ecommerce_data.csv`.


```python
import pandas as pd
import numpy as np
```

```python
# Load dataset

df = pd.read_csv('datasets/ecommerce_data.csv')
print('Shape:', df.shape)
df.head()
```

## 1) Range of customer age


```python
age_min = int(df['age'].min())
age_max = int(df['age'].max())
age_range = age_max - age_min

print('Minimum age:', age_min)
print('Maximum age:', age_max)
print('Age range:', age_range)
```

## 2) Mean, median, and mode of amount spent


```python
amount_mean = float(df['amount_spend'].mean())
amount_median = float(df['amount_spend'].median())
amount_mode = df['amount_spend'].mode().tolist()

print('Mean amount spent:', round(amount_mean, 2))
print('Median amount spent:', round(amount_median, 2))
print('Mode amount spent:', amount_mode)
```

## 3) Variance and standard deviation of time spent on site


```python
# Sample variance and standard deviation (ddof=1)
time_variance = float(df['time_spent_on_site'].var(ddof=1))
time_std = float(df['time_spent_on_site'].std(ddof=1))

print('Variance (sample):', round(time_variance, 2))
print('Standard deviation (sample):', round(time_std, 2))
```

## 4) Interquartile range (IQR) of amount spent


```python
q1 = float(df['amount_spend'].quantile(0.25))
q3 = float(df['amount_spend'].quantile(0.75))
iqr = q3 - q1

print('Q1 (25th percentile):', round(q1, 2))
print('Q3 (75th percentile):', round(q3, 2))
print('IQR:', round(iqr, 2))
```

## 5) Correlation between number of purchases and time spent on site


```python
corr_purchases_time = float(df['no_of_purchases'].corr(df['time_spent_on_site']))

print('Correlation coefficient:', round(corr_purchases_time, 2))
```

## 6) Z-score for a customer who spends $450


```python
x = 450
z_score_450 = (x - amount_mean) / df['amount_spend'].std(ddof=1)

print('Z-score for amount_spend = $450:', round(float(z_score_450), 2))
```

## 7) Skewness of amount spent distribution


```python
amount_skewness = float(df['amount_spend'].skew())

if amount_skewness > 0:
    skew_interpretation = 'Right-skewed (positive skew)'
elif amount_skewness < 0:
    skew_interpretation = 'Left-skewed (negative skew)'
else:
    skew_interpretation = 'Approximately symmetric'

print('Skewness:', round(amount_skewness, 2))
print('Interpretation:', skew_interpretation)
```

## 8) Probability of more than 5 purchases OR spending more than $300


```python
A = df['no_of_purchases'] > 5
B = df['amount_spend'] > 300

p_A = A.mean()
p_B = B.mean()
p_A_and_B = (A & B).mean()
p_A_or_B = p_A + p_B - p_A_and_B

print('P(A):', round(float(p_A), 3))
print('P(B):', round(float(p_B), 3))
print('P(A and B):', round(float(p_A_and_B), 3))
print('P(A or B):', round(float(p_A_or_B), 3))
```

## 9) 95% confidence interval for amount spent


```python
n = len(df)
amount_std = float(df['amount_spend'].std(ddof=1))
z_critical = 1.96
margin_of_error = z_critical * (amount_std / np.sqrt(n))
ci_lower = amount_mean - margin_of_error
ci_upper = amount_mean + margin_of_error
print('Sample size:', n)
print('Mean amount spent:', round(amount_mean, 2))
ci_95 = (float(round(ci_lower)), float(round(ci_upper)))
print('95% CI:', ci_95)
```

## 10) Conditional probability: high cross-sell given 5+ purchases


```python
condition_purchases_5_plus = df['no_of_purchases'] >= 5
condition_high_cross_sell = df['cross_sell_conversion_rate'] > 0.80

p_high_cross_sell_given_5_plus = (
    (condition_purchases_5_plus & condition_high_cross_sell).sum()
    / condition_purchases_5_plus.sum()
)

print('P(cross_sell_conversion_rate > 0.80 | no_of_purchases >= 5):',
      round(float(p_high_cross_sell_given_5_plus), 3))
```
